```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
path = "/Users/amrtamer/Documents/Internship"
# path = "/content/drive/MyDrive/Internship"
```

```python
data = pd.read_csv(path + r'/data/Volve/Volve_cleaned.csv')
data.head()
```

```python
well_info = dict(set([(j, i) for i, j in data[['WELL_BORE_CODE', 'well_name']].values.tolist()])) # creating a dictionary of well identiification with {code: name} structure
well_info
```

```python
identification_column = ['well_name', 'WELL_BORE_CODE', 'date']
feature_columns = ['prod_hrs', 'bhp', 'bht', 'dp_tubing', 'AVG_ANNULUS_PRESS', 'AVG_CHOKE_SIZE_P', 'whp', 'wht', 'choke_size_percentage', 'oil_vol', 'gas_vol', 'water_vol', 'oil_rate']
target_column = ['oil_rate']
```

```python
data = data[identification_column + feature_columns]
data.shape
```

```python
features = data[feature_columns]
data[feature_columns] = (features - features.mean()) / features.std()
data
```

```python
# Initialize an empty list to store predictions
data['linear_regressor'] = None
for well_name in well_info.keys():
    rolling_predictions = []
    well = data[data['well_name'] == well_name].reset_index(drop=True)
    well_index = data[data['well_name'] == well_name].index
    # Loop over each time step in the dataset
    for i in range(len(well)):
        if i > 0:
            X_train = well[feature_columns].iloc[:i]
            Y_train = well[target_column].iloc[:i]

            # Initialize and fit the linear regression model
            linear_regressor = LinearRegression()
            linear_regressor.fit(X_train, Y_train)

            # Predict the target value for the current data point
            current_features = well[feature_columns].iloc[[i]]
            prediction = linear_regressor.predict(current_features)[0]
        else:
            # If not enough data points, use a default value (e.g., the target mean or previous value)
            prediction = well[target_column].iloc[0]
        # Append the prediction to the list
        rolling_predictions.append(prediction)
    data.loc[well_index, 'linear_regressor'] = rolling_predictions
data
```

```python
# Compute EMA with different spans
span_short = 5
span_medium = 10
span_long = 20

# Initialize an empty list to store predictions
data['EMA_short'] = None
data['EMA_medium'] = None
data['EMA_long'] = None
for well_name in well_info.keys():
    rolling_predictions = []
    well = data[data['well_name'] == well_name].reset_index(drop=True)
    well_index = data[data['well_name'] == well_name].index
    ema_short = well[target_column].ewm(span=span_short, adjust=False).mean().values.reshape(-1)
    ema_medium = well[target_column].ewm(span=span_medium, adjust=False).mean().values.reshape(-1)
    ema_long = well[target_column].ewm(span=span_long, adjust=False).mean().values.reshape(-1)
    data.loc[well_index, 'EMA_short'] = ema_short
    data.loc[well_index, 'EMA_medium'] = ema_medium
    data.loc[well_index, 'EMA_long'] = ema_long
data
```

```python
# Rolling window spans
window_short = 5
window_medium = 10
window_long = 20

# Initialize the rolling window columns in the DataFrame with NaN
data['Rolling_short'] = None
data['Rolling_medium'] = None
data['Rolling_long'] = None

# Iterate over each well
for well_name in data['well_name'].unique():
    # Filter the DataFrame for the current well
    well = data[data['well_name'] == well_name].reset_index(drop=True)
    well_index = data[data['well_name'] == well_name].index
    
    # Compute the rolling mean for each window
    rolling_short = well[target_column].rolling(window=window_short).mean().fillna(0).values.reshape(-1)
    rolling_medium = well[target_column].rolling(window=window_medium).mean().fillna(0).values.reshape(-1)
    rolling_long = well[target_column].rolling(window=window_long).mean().fillna(0).values.reshape(-1)

    # Store the rolling means back in the original DataFrame
    data.loc[well_index, 'Rolling_short'] = rolling_short
    data.loc[well_index, 'Rolling_medium'] = rolling_medium
    data.loc[well_index, 'Rolling_long'] = rolling_long
data
```

```python
short_lag = 3
medium_lag = 5
long_lag = 10

data['Lag_short'] = None
data['Lag_medium'] = None
data['Lag_long'] = None

# Iterate over each well
for well_name in well_info.keys():
    # Filter the DataFrame for the current well
    well = data[data['well_name'] == well_name].reset_index(drop=True)
    well_index = data[data['well_name'] == well_name].index
    for lag, name in zip([short_lag, medium_lag, long_lag], ['Lag_short', 'Lag_medium', 'Lag_long']):
        lags = []
        # Create lagged values with padding for the initial values
        start_shift = well[target_column].shift(lag).fillna(0).values.reshape(-1)
        for i in range(0, lag): 
            lags.append(list(start_shift[i:i+lag]))

        # Create lagged values for the remaining rows
        for i in range(lag-1, len(well) - 1): 
            lags.append(list(well.loc[i-lag+1:i, target_column].values.reshape(-1)))

        # Assign the lagged list to the DataFrame
        for i, j in enumerate(list(well_index)):
            data.at[j, name] = lags[i]

data
```

```python
new_feature_columns = ['linear_regressor',	'EMA_short', 'EMA_medium', 'EMA_long', 'Rolling_short',	'Rolling_medium', 'Rolling_long']
new_features = data[new_feature_columns]
data[new_feature_columns] = (new_features - new_features.mean()) / new_features.std()
data
```
